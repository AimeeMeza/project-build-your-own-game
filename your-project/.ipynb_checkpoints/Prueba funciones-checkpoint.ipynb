```python
#Lanza una palabra al azar
import random
def palabra_azar(tema,dificultad):
    
    Astronomia4=['luna','nasa','star','roja']
    Astronomia5=['marte','venus','space','urano']
    Astronomia10=['planetoide','telescopio','astronomia','planetaria']
    Biologia4=['agua','raiz','alga','aves']
    Biologia5=['cinco','genes','pulpo','golgi']
    Biologia10=['sanguineos','nutrientes','citoplasma','aminoacido']

    #Astronomia
    if tema==1 and dificultad==1:
        palabra=random.choice(Astronomia4)
        return palabra
    elif tema==1 and dificultad==2:
        palabra=random.choice(Astronomia5)
        return palabra
    elif tema==1 and dificultad==3:
        palabra=random.choice(Astronomia10)
        return palabra
    #Biologia
    elif tema==2 and dificultad==1:
        palabra=random.choice(Biologia4)
        return palabra
    elif tema==2 and dificultad==2:
        palabra=random.choice(Biologia5)
        return palabra
    elif tema==2 and dificultad==3:
        palabra=random.choice(Biologia10)
        return palabra
    #Random
    elif tema==3 and dificultad==1:
        Random4=Astronomia4+Biologia4
        palabra=random.choice(Random4)
        return palabra
    elif tema==3 and dificultad==2:
        Random5=Astronomia5+Biologia5
        palabra=random.choice(Random5)
        return palabra
    elif tema==3 and dificultad==3:
        Random10=Astronomia10+Biologia10
        palabra=random.choice(Random10)
        return palabra
```

```python
print(palabra_azar(3,2))
```

```python
#Manda la pista de la palabra, esta funcion llama la funcion palabra_azar
def pistas(tema,dificultad):
  
    palabra=palabra_azar(tema,dificultad)
    
    #Pistas para palabras tema 1 dificultad 1
    if palabra == 'luna':
        Frase=('Quinto satelite más grande del sistema solar') 
    elif palabra == 'star':
        Frase= ('Termino en ingles de objeto unico en el sistema solar')
    elif palabra=='nasa':
        Frase=('Su sede se encuentra el Washington')
    elif palabra=='roja':
        Frase=('La gigante___ es una estrella grande y fria')

#     #Pistas para palabras tema 1 dificultad 2
    elif palabra == 'marte':
         Frase=('Segundo planeta más pequeño del Sistema solar') 
    elif palabra == 'venus':
         Frase= ('Planeta rocoso con atmosfera de dióxido de carbono ')
    elif palabra=='space':
         Frase=('Johnson ______ Center')
    elif palabra=='urano':
         Frase=('Descubierto por primera vez con ayuda de un telescopio')
            
     #Pistas para palabras tema 1 dificultad 3
    elif palabra == 'planetoide':
         Frase=('Sinonimo de asteroide') 
    elif palabra == 'telescopio':
         Frase= ('Hubble')
    elif palabra=='astronomia':
         Frase=('trata de magnitud, medida y movimiento ')
    elif palabra=='planetaria':
         Frase=('es un tipo de nebulosa')
        
         #Pistas para palabras tema 2 dificultad 1
    elif palabra == 'agua':
         Frase=('clave para la vida') 
    elif palabra == 'raiz':
         Frase= ('lo primero en la germinación')
    elif palabra=='alga':
         Frase=('Planta primitiva')
    elif palabra=='aves':
         Frase=('Su antecesor pudieron ser los dinosaurios')
     #Pistas para palabras tema 2 dificultad 2
    elif palabra == 'cinco':
         Frase=('Reinos de clasificación de los seres vivos') 
    elif palabra == 'genes':
         Frase= ('segemtnos de ADN ')
    elif palabra=='pulpo':
         Frase=('invertebrado')
    elif palabra=='golgi':
         Frase=('El aparato de ____ solo esta en celulas eucariontas')
#     #Pistas para palabras tema 2 dificultad  3
    elif palabra == 'sanguineos':
         Frase=('los vasos____ se clasifican en 5 grupos') 
    elif palabra == 'nutrientes':
         Frase= ('Ayuda a las celulas con sus funciones vitales')
    elif palabra=='citoplasma':
         Frase=('parte de celula')
    elif palabra=='aminoacido':
         Frase=('molecula organica')
        
    return palabra,Frase
```

```python
#Funcion principal

def main():
    #Comienzo del juego; el usuario selecciona modo de juego
    jugadores=int(input('Selecciona el numero de jugadores escribe 1 ó 2:'))
    rondas=int(input('Selecciona el numero de rondas escribe 1 ó 2:'))
    tema=int(input('Selecciona la tematica de juego : Astronomía = 1, Biología =2, Random= 3'))
    dificultad=int(input('Selecciona la dificultad del juego: baja: 1 ,media: 2,alta: 3'))

    #Variables para ciclos y modo de juego
    rondas_completadas=0
    jugador=0
    
    #Crear primer ciclo para numero de jugadores, parara hasta que el numero de jugadores sea igual al seleccionado:
    while jugador < jugadores:
    #Segundo ciclo , finaliza hasta que el numero de rondas sea = al numero de rondaas seleccionadas
        while rondas_completadas < rondas:
            
            palabra,Frase=pistas(tema,dificultad)

            tablero(palabra,Frase)
            
            rondas_completadas+=1

            jugador+=1
```

```python
main()
```
